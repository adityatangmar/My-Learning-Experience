```python
import pandas as pd
df = pd.read_csv('https://drive.google.com/uc?id=1XZbZ5Z7MBXYaR3UO1HbQ4ncPi4R3HsyF')
df.head()
```

```python
df.info()
```

```python
# Map names to numeric labels safely
# grapefruit -> 0, orange -> 1; other names -> -1
df['label'] = df['name'].map({'grapefruit': 0, 'orange': 1})
df['label'] = df['label'].fillna(-1).astype(int)
```

```python
from sklearn import preprocessing

# Encode the 'name' column using LabelEncoder
label_encoder = preprocessing.LabelEncoder()
df['name'] = label_encoder.fit_transform(df['name'])
print(df)
```

```python
df.info()
```

```python
# mengembalikan dataset dalam bentuk array numpy
dataset = df.values
```

```python
'''pilih 5 kolom terakhir sebagai atribut
bilangan sebelum koma untuk memilih baris pada dataframe
bilangan setelah koma untuk memilih kolom pada dataframe'''
X = dataset[:,1:6]
```

```python
# memilih kolom pertama sebagai label
y = dataset[:,0]
```

```python
from sklearn import preprocessing

# normalisasi data dengan MinMaxScaler
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
X_scale
```

```python
# pisahkan data training dan testing
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_scale, y, test_size=0.3)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# buat model Sequential
model = Sequential([
                    Dense(32, activation='relu', input_shape=(5,)),
                    Dense(32, activation='relu'),
                    Dense(1, activation='sigmoid')])
```

```python
# kompilasi model
model.compile(optimizer='sgd',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# latih model
model.fit(X_train, Y_train, epochs=100)
```

```python
# evaluasi model
model.evaluate(X_test, Y_test, batch_size=1)
```
